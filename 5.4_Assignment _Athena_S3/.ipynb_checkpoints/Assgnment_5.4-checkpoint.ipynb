```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_parquet('earnings_date=2022-02-10')
df.info()
```

```python
df
```

```python
def create_data_frames():
    min_value = df["earnings"].min()
    avg_value = df["earnings"].max()
    df['earnings'] = np.random.randint(min_value, avg_value, size=df.shape[0])
    return df
    
```

```python
print(create_data_frames())
```

```python

for i in range(2):
    table=create_data_frames()
    table.to_parquet("earnings_data{}".format(i))
  
    
    
```
